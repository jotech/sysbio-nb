```python
from pyndamics import Simulation
```

```python
sim = Simulation()

sim.add("firs'   = -cattles",       1, plot=True)
sim.add("cattles'= insects",        1, plot=True)
sim.add("plants' = firs",           1, plot=True)
sim.add("insects'= plants - birds", 1, plot=True)
sim.add("birds'  = insects - hawks",1, plot=True)
sim.add("hawks'  = insects",        1, plot=True)

#sim.params(c=0.01)

sim.run(0,5)
```
